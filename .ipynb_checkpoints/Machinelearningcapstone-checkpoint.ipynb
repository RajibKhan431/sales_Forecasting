 Build and compare linear regression, random forest, and XGBoost models for predictions

```python
final_df.head()
```

```python
final_df['day'] = final_df['date'].dt.day
```

```python
from sklearn.preprocessing import LabelEncoder
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

le = LabelEncoder()
le.fit(order)

final_df['day_of_week_encoded'] = le.transform(final_df['day_of_week'])
```

```python
final_df.head()
```

```python
features = [
    'day', 'month', 'year',
    'day_of_week_encoded', 'quarter',
    'price'

]
X = final_df[features]
y = final_df['item_count']
```

```python
final_df = final_df.sort_values('date')
X = X.loc[final_df.index]
y = y.loc[final_df.index]
```

```python
#using the data of last six month
```

```python
test_size = 6  # months
split_date = final_df['date'].max() - pd.DateOffset(months=test_size)

X_train = X[final_df['date'] < split_date]
X_test  = X[final_df['date'] >= split_date]

y_train = y[final_df['date'] < split_date]
y_test  = y[final_df['date'] >= split_date]
```

```python
## linear regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
```

```python
## random forest
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

```python
#install XGBoost using pip
```

```python
%pip install xgboost --user
```

```python
# XGBoost
from xgboost import XGBRegressor
xgb_model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=42,
    objective='reg:squarederror'
)

xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
```

```python
#Computing the root mean square error (RMSE) values for each model to compare their performances 
```

```python
from sklearn.metrics import mean_squared_error

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_lr  = rmse(y_test, lr_pred)
rmse_rf  = rmse(y_test, rf_pred)
rmse_xgb = rmse(y_test, xgb_pred)


print("Linear Regression RMSE :", rmse_lr)
print("Random Forest RMSE     :", rmse_rf)
print("XGBoost RMSE           :", rmse_xgb)
```

```python
##compare model which has less RMSE value
rmse_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest','XGBoost'],
    'RMSE': [rmse_lr, rmse_rf,rmse_xgb]
})

rmse_df.sort_values('RMSE')
```

```python
#Best model is random forest because it has less RMSE value.
```

```python
best_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

best_model.fit(X, y)

```

```python
import pickle

pickle.dump(best_model, open("model.pkl", "wb"))

pickle.dump(le, open("label_encoder.pkl", "wb"))

print("Model Saved Successfully!")
```

```python
#Use the best-performing models to make a forecast for the next year
```

```python
# Generate next 12 months forecasting sale
last_date = final_df['date'].max()

future_dates = pd.date_range(
    start=last_date + pd.offsets.MonthEnd(1),
    periods=12,
    freq='ME'
)

future_df = pd.DataFrame({'date': future_dates})

# Feature engineering (same as training)
future_df['day'] = future_df['date'].dt.day
future_df['month'] = future_df['date'].dt.month
future_df['year'] = future_df['date'].dt.year
future_df['quarter'] = future_df['date'].dt.quarter
future_df['day_of_week'] = future_df['date'].dt.day_name()

# Encode day of week
future_df['day_of_week_encoded'] = le.transform(future_df['day_of_week'])

# Price assumption 
future_df['price'] = final_df['price'].mean()
```

```python
future_X = future_df[features]

future_df['forecasted_item_count'] = best_model.predict(future_X)

future_df[['date', 'forecasted_item_count']]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(future_df['date'], future_df['forecasted_item_count'], 
         marker='o', linestyle='-', color='b', label='Forecast')

plt.title('Forecasted Item Count for Next 12 Months')
plt.xlabel('Date')
plt.ylabel('Forecasted Item Count')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
